```python
results = [
    './results_05_dev1024_multimodel/sub_a0961_bert-base-uncased_3e_v1024o_210319_2253.csv',
    './results_05_dev1024_multimodel/sub_a0954_distilbert-base-uncased_3e_v1024o_210319_2234.csv',
    './results_05_dev1024_multimodel/sub_a0887_roberta-base_2e_v1024o_210319_2201.csv'
]
```

```python
rdfs = []
for result in results:
    rdf = pd.read_csv(result)
    rdfs.append(rdf)
```

```python
for rdf in rdfs:
    print(rdf.head())
```

```python
ensemble_df = rdfs[0]
```

```python
ensemble_df['target_1'] = rdfs[1]['target']
ensemble_df['target_2'] = rdfs[2]['target']
```

```python
ensemble_df.head(20)
```

```python
ensemble_df['target_ens'] = ensemble_df \
    .apply(lambda r: 1 if r[['target', 'target_1', 'target_2']].sum() >= 2 else 0, axis=1)
```
